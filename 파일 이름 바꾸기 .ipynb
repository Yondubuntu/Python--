```python
import os

path = "./filepath/"
def changeName(path):
    i = 1
    for filename in os.listdir(path):
        print(path+filename, '=>', path+str(i)+'.jpg')
        os.rename(path+filename, path+str(i)+'.jpg')
        i += 1


changeName('filepath')


```
